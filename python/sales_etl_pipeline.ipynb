```python
import pandas as pd
```

```python
df = pd.read_excel("raw_sales_data.xlsx")

df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("%", "percent")
)
```

```python
df.columns
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df["revenue"] = (
    df["revenue"]
    .replace("[^0-9.]", "", regex=True)
    .astype(float)
)

df["discount"] = (
    df["discount"]
    .replace("[^0-9.]", "", regex=True)
    .astype(float)
)
```

```python
df["net_revenue"] = df["revenue"] - df["discount"]
```

```python
df["month"] = df["order_date"].dt.month
df["year"] = df["order_date"].dt.year
```

```python
df["margin_percent"] = (df["net_revenue"] / df["revenue"]) * 100
```

```python
df = df.dropna()
```

```python
df["order_date"] = df["order_date"].dt.strftime("%Y-%m-%d %H:%M:%S")
```

```python
df = df[[
    "order_id",
    "order_date",
    "region",
    "sales_rep",
    "product_category",
    "revenue",
    "discount",
    "net_revenue",
    "month",
    "year",
    "margin_percent"
]]
```

```python
df.to_csv("sales_data_postgres_ready.csv", index=False)
```
